```python
# Path folder data asli
base_image_folder = "D:/Ando File 4 Kuliah/A SKRIPSI/RISETku/KERJA/DATA/images/train/"

# Path folder untuk menyimpan gambar grayscale
base_save_folder = "D:/Ando File 4 Kuliah/A SKRIPSI/RISETku/KERJA/DATA/images/grayscale/"

# Membuat folder grayscale jika belum ada
if not os.path.exists(base_save_folder):
    os.makedirs(base_save_folder)

# Loop melalui setiap folder kategori (Angry, Disgust, Fear, dll.)
for category in os.listdir(base_image_folder):
    category_path = os.path.join(base_image_folder, category)
    
    # Mengecek apakah path adalah sebuah folder
    if os.path.isdir(category_path):
        # Membuat folder kategori di folder grayscale jika belum ada
        save_category_folder = os.path.join(base_save_folder, category)
        if not os.path.exists(save_category_folder):
            os.makedirs(save_category_folder)
        
        # Mendapatkan semua file gambar dalam kategori
        image_files = [f for f in os.listdir(category_path) if f.endswith(".png") or f.endswith(".jpg")]
        
        # Menampilkan progress bar untuk setiap kategori
        print(f"Processing category: {category}")
        for filename in tqdm(image_files, desc=f"{category} Progress"):
            img_path = os.path.join(category_path, filename)
            img = Image.open(img_path)

            # Mengonversi gambar ke grayscale
            gray_img = img.convert('L')

            # Menyimpan gambar grayscale ke folder kategori yang sesuai
            save_path = os.path.join(save_category_folder, filename)
            gray_img.save(save_path)

print("Semua gambar berhasil dikonversi menjadi grayscale.")
```

The Specific operation contain 4 convolution layers:
-  Patch Extraction and Representation

























    $$F_1(P) = \max(0, W_1 + P + B_1)$$

    Where 𝑃 P is an input image patch with size 𝑛 × 𝑛 n×n, 𝑊 1 W 1 ​ and 𝐵 1 B 1 ​ are the weights and biases of the filters. The size of 𝑊 1 W 1 ​ is 𝑓 1 × 𝑓 1 × 𝑛 1 f 1 ​ ×f 1 ​ ×n 1 ​ , where 𝑓 1 f 1 ​ is the spatial support of a filter, and 𝑛 1 n 1 ​ is the number of filters. 𝐵 1 B 1 ​ is an 𝑛 1 n 1 ​ -dimensional vector, with each element associated with a filter. The symbol “∗” represents the convolution operation. Max(0, x) is a Rectified Linear Unit (ReLU) activation function (Krizhevsky et al., 2012), which is applied to accelerate training convergence and improve performance.

-  Feature Enhancement

    $$F_2(P) = \max(0, W_2 * F_1(P) + B_2)$$

    Where 𝑊 2 W 2 ​ contains 𝑛 2 n 2 ​ filters of size 𝑓 2 × 𝑓 2 × 𝑛 1 f 2 ​ ×f 2 ​ ×n 1 ​ , and 𝐵 2 B 2 ​ is an 𝑛 2 n 2 ​ -dimensional vector.

-  Non-LinearMapping

    $$F_3(P) = \max(0, W_3 * F_2(P) + B_3)$$

    Where 𝑊 3 W 3 ​ contains 𝑛 3 n 3 ​ filters of size 𝑓 3 × 𝑓 3 × 𝑛 2 f 3 ​ ×f 3 ​ ×n 2 ​ , and 𝐵 3 B 3 ​ is an 𝑛 3 n 3 ​ -dimensional vector.

-  Reconstruction

    $$F_4(P) = W_4 * F_3(P) + B_4$$

    Where 𝑊 4 W 4 ​ contains 𝑛 4 n 4 ​ filters of size 𝑓 4 × 𝑓 4 × 𝑛 3 f 4 ​ ×f 4 ​ ×n 3 ​ , and 𝐵 4 B 4 ​ is an 𝑛 4 n 4 ​ -dimensional vector.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class LightenNet(nn.Module):
    def __init__(self):
        super(LightenNet, self).__init__()
        
        # Patch Extraction and Representation (32 filters, 6x6 kernel, no padding)
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=6, stride=1, padding=0)
        
        # Feature Enhancement (16 filters, 7x7 kernel, no padding)
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=16, kernel_size=7, stride=1, padding=0)
        
        # Non-Linear Mapping (8 filters, 1x1 kernel, no padding)
        self.conv3 = nn.Conv2d(in_channels=16, out_channels=8, kernel_size=1, stride=1, padding=0)
        
        # Reconstruction (1 filter, 5x5 kernel, no padding)
        self.conv4 = nn.Conv2d(in_channels=8, out_channels=1, kernel_size=5, stride=1, padding=0)

    def forward(self, x):
        # Patch Extraction and Representation
        x = F.relu(self.conv1(x))  # Output size will be 32 x 11 x 11
        
        # Feature Enhancement
        x = F.relu(self.conv2(x))  # Output size will be 16 x 5 x 5
        
        # Non-Linear Mapping
        x = F.relu(self.conv3(x))  # Output size will be 8 x 5 x 5
        
        # Reconstruction
        x = self.conv4(x)  # Output size will be 1 x 1 x 1
        
        return x
```

```python
# Load model LightenNet
model = LightenNet()
model.eval()  # Set model to evaluation mode

# Transformasi dataset SFEW
transform = transforms.Compose([
    transforms.Resize((16, 16)),  # Sesuaikan input dengan ukuran yang diinginkan
    transforms.ToTensor(),
])

# Load dataset SFEW
dataset = datasets.ImageFolder(root=r"D:/Ando File 4 Kuliah/BELAJAR SKRIPSI/RISETku/KERJA/DATA/images/train", transform=transform)
dataloader = torch.utils.data.DataLoader(dataset, batch_size=1, shuffle=False)

# Path untuk menyimpan output
output_folder = 'D:/Ando File 4 Kuliah/BELAJAR SKRIPSI/RISETku/KERJA/DATA/images/outputLight'
os.makedirs(output_folder, exist_ok=True)

# Fungsi untuk menyimpan output sebagai gambar
def save_output_image(output_tensor, output_path):
    # Pastikan tensor memiliki dimensi (1, 1, H, W)
    output_tensor = output_tensor.squeeze(0)  # Menghapus batch dimension, hasil (1, H, W)
    
    if output_tensor.dim() == 3:
        output_tensor = output_tensor.squeeze(0)  # Menghapus channel dimension, hasil (H, W)
    
    # Sekarang harus memiliki dua dimensi (H, W)
    if output_tensor.dim() == 2:
        output_image = output_tensor.detach().cpu().numpy()
        output_image = Image.fromarray((output_image * 255).astype('uint8'), mode='L')
        output_image.save(output_path)
    else:
        print(f"Unexpected output shape: {output_tensor.shape}")



# Iterasi untuk setiap batch dalam dataloader
for i, (images, _) in enumerate(dataloader):
    with torch.no_grad():
        # Forward pass
        output = model(images)
        
        # Simpan output ke file
        output_path = os.path.join(output_folder, f"output_{i}.png")
        save_output_image(output, output_path)

    print(f"Saved output {i} to {output_path}")
```

# CLAHE

```python
import cv2
import matplotlib.pyplot as plt

# Membaca gambar dalam mode grayscale
img = cv2.imread("D:/Ando File 4 Kuliah/BELAJAR SKRIPSI/RISETku/KERJA/DATA/images/train/Angry/RevolutionaryRoad_000654280_00000045.png", cv2.IMREAD_GRAYSCALE)

# Membuat objek CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# Melakukan CLAHE pada gambar
clahe_img = clahe.apply(img)

# Menyimpan hasil gambar
cv2.imwrite('gambar_CLAHE.jpg', clahe_img)  # Menyimpan hasil CLAHE

# Menampilkan gambar asli dan hasilnya
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(clahe_img, cmap='gray')
plt.title('Gambar CLAHE')
plt.axis('off')

plt.show()

# Menyimpan gambar grayscale
print("Gambar grayscale berhasil disimpan sebagai 'grayscale_image.png'")
```
